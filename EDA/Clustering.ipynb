## Clustering

```python
from sklearn.cluster import KMeans
kmc1 = KMeans(n_clusters = 3, random_state = 2021)
kmc1.fit(data)
```

```python
c_data = data.copy()
c_data['Cluster'] = kmc1.fit_predict(c_data)
```

```python
c_data['color'] = np.where(c_data.Cluster == 0, 'skyblue', 

                         np.where(c_data.Cluster == 1, 

                         'yellowgreen', 

                         'green'))
```

## PCA

```python
from sklearn.preprocessing import StandardScaler  # 표준화 패키지 라이브러리 
x = data.values       # 독립변인들의 value값만 추출
target =  kmc1.fit_predict(c_data)              # 종속변인 추출

X_train_scaled = StandardScaler().fit_transform(x) # x객체에 x를 표준화한 데이터를 저장
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 주성분을 몇개로 할지 결정
printcipalComponents = pca.fit_transform(X_train_scaled )
principalDf = pd.DataFrame(data=printcipalComponents, columns = ['principal component1', 'principal component2'])
# 주성분으로 이루어진 데이터 프레임 구성

# X와 target을 붙인 finalDf 생성
finalDf = pd.concat([principalDf, pd.Series(target, name='target')], axis = 1)
finalDf.rename(columns={0: "target"})
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize=20)

targets = [0,1,2]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component1']
               , finalDf.loc[indicesToKeep, 'principal component2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid(True)
```
